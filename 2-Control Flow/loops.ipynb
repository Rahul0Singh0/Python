```python
## for loop - iterate over a sequence of numbers
for i in range(5):
    print(i)
```

```python
for i in range(3, 10):
    print(i)
```

```python
range(1, 10, 2)
```

```python
for i in range(1, 20, 2): ## increament by 2
    print(i)
```

```python
for i in range(1, 20, -2):
    print(i)
```

```python
## strings

str = "Rahul"
for i in str:
    print(i)
```

```python
## while loop 
## it continues to execute as long as the condition is True.

count = 0

while count < 5:
    print(count)
    count = count + 1
```

```python
## Loop control statement

## break
## The break statement exists the loop permaturely

## break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue

## The continue statement skiips the current iteration and continues with the next

for i in range(10):
    if i == 3:
        continue
    print(i)
```

```python
## pass 
## The pass statement is a null operation; it does nothing 

for i in range(5):
    if i == 3:
        pass
    print(i)
```
