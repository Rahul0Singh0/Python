```python
## if statement
age = 20
if age >= 20:
    print("Eligible for vote")
```

```python
## else
## the else statement executes a block of code if the condition in the if statement is False.

age == 16

if age >= 18:
    print("eligible")
else: 
    print("not eligible, you are minor")
```

```python
## elif
## It allows to you to check multiple conditions. It stands for "else if"4

if age < 13:
    print("you are a child")
elif age < 18: 
    print("you are teenager")
else:
    print("you are eligible")
```
