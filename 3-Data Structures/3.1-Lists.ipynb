```python
## Accessing list elements by its index

fruits = ["apple", "banana", "cherry", "kiwi", "pineapple"]

print(fruits[0])
print(fruits[2])
print(fruits[1])
print(fruits[-1])
```

```python
print(fruits[1:])
print(fruits[1:3])
print(fruits[-1:]) ## from -1 index to last index

print(fruits[-3:-1])
```

```python
fruits[1] = "watermelon" ## mutable
print(fruits)
```

```python
fruits[1:] = "Watermelon"
print(fruits)
```
